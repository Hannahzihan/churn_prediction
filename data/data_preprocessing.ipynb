Transaction features extraction (during 205-2016)

```python
import pandas as pd

transactions = pd.read_csv(r"data_intermediate/transactions_all.csv")
```

```python
# Transactions data

transactions["transaction_date"] = pd.to_datetime(transactions["transaction_date"], format="%Y%m%d")
transactions["membership_expire_date"] = pd.to_datetime(transactions["membership_expire_date"], format="%Y%m%d")

transaction_count = transactions.groupby('msno')['payment_method_id'].count().rename('transaction_count')

is_auto_renew_change_flag = transactions.groupby('msno')['is_auto_renew'].max().rename('is_auto_renew_change_flag')
transactions['autorenew_not_cancel'] = ((transactions['is_auto_renew'] == 1) & (transactions['is_cancel'] == 0)).astype(int)
autorenew_not_cancel = transactions.groupby('msno')['autorenew_not_cancel'].min().rename('autorenew_not_cancel')

average_plan = transactions.groupby('msno')['payment_plan_days'].mean().rename('average_plan')
change_in_method = transactions.groupby('msno')['payment_method_id'].nunique().rename('change_in_method')
change_in_plan = transactions.groupby('msno')['payment_plan_days'].nunique().rename('change_in_plan')
sum_plan_days = transactions.groupby('msno')['payment_plan_days'].sum().rename('sum_plan_days')

average_amount_paid = transactions.groupby('msno')['actual_amount_paid'].mean().rename('average_amount_paid')
average_amount_charged = transactions.groupby('msno')['plan_list_price'].mean().rename('average_amount_charged')
difference = (average_amount_paid - average_amount_charged).rename("difference")
change_in_price = transactions.groupby('msno')['plan_list_price'].nunique().rename('change_in_price')
change_in_price_paid = transactions.groupby('msno')['actual_amount_paid'].nunique().rename('change_in_price_paid')
sum_list_price = transactions.groupby('msno')['plan_list_price'].sum().rename('sum_list_price')
sum_amount_paid = transactions.groupby('msno')['actual_amount_paid'].sum().rename('sum_amount_paid')


duration = (transactions['membership_expire_date'] - transactions['transaction_date']).dt.days.groupby(transactions['msno']).mean().rename('duration')

more_than_30 = (transactions['membership_expire_date'] - transactions['transaction_date']).dt.days.groupby(transactions['msno']).apply(lambda x: (x > 30).sum()).rename('more_than_30')

features = [
    transaction_count, is_auto_renew_change_flag,
    average_plan, change_in_method, change_in_plan, sum_plan_days,
    average_amount_paid, average_amount_charged, difference, change_in_price,
    change_in_price_paid, sum_list_price, sum_amount_paid, duration, more_than_30,
    autorenew_not_cancel
]

transaction_features = pd.concat(features, axis=1)

file_path = r"data_feature/transaction_features.csv"
transaction_features.to_csv(file_path, index=True)
```

Member data preparation

```python
# Members data

members = pd.read_csv(r"data_raw/members_v3.csv")

# Handle missing values and outliers

# Remove age outliers (valid range: 0-100)
members["bd"] = members["bd"].apply(lambda x: np.nan if x <= 0 or x >= 100 else x)

# Fill missing age with median
members["bd"].fillna(members["bd"].median(), inplace=True)

# Fill missing categorical values with mode
members["gender"].fillna("Unknown", inplace=True)

# Convert date columns to datetime format
members["registration_init_time"] = pd.to_datetime(members["registration_init_time"], format="%Y%m%d", errors="coerce")

# Define reference date
reference_date = pd.to_datetime("2017-02-01")

# Compute registration duration in days
members["registration_duration_days"] = (reference_date - members["registration_init_time"]).dt.days

members.to_csv(r"data_intermediate/members.csv",index=False)
```

```python
import pandas as pd
labeled_data = pd.read_csv(r"data_raw/train_v2.csv")
user_logs_jan = pd.read_csv(r"data_feature/jan_log_user_features.csv")
user_logs_feb = pd.read_csv(r"data_feature/feb_log_user_features.csv")
transaction_features = pd.read_csv(r"data_feature/transaction_features.csv")
members = pd.read_csv(r"data_intermediate/members.csv")
```

```python
# merge and fill na

labeled_data_users = labeled_data["msno"]
user_logs = pd.merge(labeled_data_users, user_logs_jan, on="msno", how="left")
user_logs = pd.merge(user_logs, user_logs_feb, on="msno", how="left")

user_logs.fillna(0, inplace=True)  # Fill numerical missing values with 0
user_logs["last_login_date_jan"].replace(0, 19700101, inplace=True)
user_logs["last_login_date_feb"].replace(0, 19700101, inplace=True)
user_logs["last_login_date_jan"] = pd.to_datetime(user_logs["last_login_date_jan"], format="%Y%m%d", errors="coerce")
user_logs["last_login_date_feb"] = pd.to_datetime(user_logs["last_login_date_feb"], format="%Y%m%d", errors="coerce")
```

```python
# columns rename

user_logs.rename(columns={
    'sum_feb_month_25': 'sum_month_25_current',
    'sum_feb_month_50': 'sum_month_50_current',
    'sum_feb_month_75': 'sum_month_75_current',
    'sum_feb_month_985': 'sum_month_985_current',
    'sum_feb_month_100': 'sum_month_100_current',
    'mean_feb_month_25': 'mean_month_25_current',
    'mean_feb_month_50': 'mean_month_50_current',
    'mean_feb_month_75': 'mean_month_75_current',
    'mean_feb_month_985': 'mean_month_985_current',
    'mean_feb_month_100': 'mean_month_100_current',
    'mean_feb_month_num_unq': 'mean_month_num_unq_current',
    'sum_feb_month_num_unq': 'sum_month_num_unq_current',
    'sum_feb_month_total_secs': 'sum_month_total_secs_current',
    'mean_feb_month_total_secs': 'mean_month_total_secs_current',
    'feb_month_login': 'month_login_current',
    'last_login_date_feb': 'last_login_date_current',

    'sum_jan_month_25': 'sum_month_25_previous',
    'sum_jan_month_50': 'sum_month_50_previous',
    'sum_jan_month_75': 'sum_month_75_previous',
    'sum_jan_month_985': 'sum_month_985_previous',
    'sum_jan_month_100': 'sum_month_100_previous',
    'mean_jan_month_25': 'mean_month_25_previous',
    'mean_jan_month_50': 'mean_month_50_previous',
    'mean_jan_month_75': 'mean_month_75_previous',
    'mean_jan_month_985': 'mean_month_985_previous',
    'mean_jan_month_100': 'mean_month_100_previous',
    'mean_jan_month_num_unq': 'mean_month_num_unq_previous',
    'sum_jan_month_num_unq': 'sum_month_num_unq_previous',
    'sum_jan_month_total_secs': 'sum_month_total_secs_previous',
    'mean_jan_month_total_secs': 'mean_month_total_secs_previous',
    'jan_month_login': 'month_login_previous',
    'last_login_date_jan': 'last_login_date_previous'
}, inplace=True)
```

```python
# diff computing

# Compute the difference (February - January) to get "change from last month"
diff_columns = [
    "sum_month_25", "sum_month_50", "sum_month_75", "sum_month_985", "sum_month_100",
    "mean_month_25", "mean_month_50", "mean_month_75", "mean_month_985", "mean_month_100",
    "mean_month_num_unq", "sum_month_num_unq", "sum_month_total_secs", "mean_month_total_secs",
    "month_login"
]

for col in diff_columns:
    user_logs[f"diff_{col}"] = user_logs[f"{col}_current"] - user_logs[f"{col}_previous"]
```

```python
# labeled_data = pd.read_csv(r"data_raw/train_v2.csv")
labeled_data = labeled_data.merge(transaction_features, on="msno", how="left")
labeled_data = labeled_data.merge(user_logs, on="msno", how="left")
labeled_data = labeled_data.merge(members, on="msno", how="left")
```

```python
# Handle missing values of members

labeled_data["gender"].fillna("Unknown", inplace=True)

# Fill missing 'registration_init_time' with a placeholder date and convert to datetime format
labeled_data["registration_init_time"].fillna(pd.to_datetime("1970-01-01").date(), inplace=True)

labeled_data["registration_duration_days"].fillna(labeled_data["registration_duration_days"].mean(), inplace=True)

labeled_data["bd"].fillna(labeled_data["bd"].mean(), inplace=True)

# Fill missing 'city' and 'registered_via' based on the most frequent value per 'bd' (age)
city_mode_by_bd = labeled_data.groupby("bd")["city"].agg(lambda x: x.mode()[0] if not x.mode().empty else "Unknown")
labeled_data["city"] = labeled_data.apply(
    lambda row: city_mode_by_bd[row["bd"]] if pd.isna(row["city"]) else row["city"],
    axis=1
)

reg_mode_by_bd = labeled_data.groupby("bd")["registered_via"].agg(lambda x: x.mode()[0] if not x.mode().empty else "Unknown")
labeled_data["registered_via"] = labeled_data.apply(
    lambda row: reg_mode_by_bd[row["bd"]] if pd.isna(row["registered_via"]) else row["registered_via"],
    axis=1
)
```

```python
# merge and fill na

unlabeled_data_users = unlabeled_data["msno"]
user_logs = pd.merge(unlabeled_data_users, user_logs_feb, on="msno", how="left")
user_logs = pd.merge(user_logs, user_logs_mar, on="msno", how="left")

user_logs.fillna(0, inplace=True)  # Fill numerical missing values with 0
user_logs["last_login_date_feb"].replace(0, 19700101, inplace=True)
user_logs["last_login_date_march"].replace(0, 19700101, inplace=True)
user_logs["last_login_date_feb"] = pd.to_datetime(user_logs["last_login_date_feb"], format="%Y%m%d", errors="coerce")
user_logs["last_login_date_march"] = pd.to_datetime(user_logs["last_login_date_march"], format="%Y%m%d", errors="coerce")
```
